```python
def predict_winner(s):
    Apples = 0
    Oranges = 0
    ans = ''
    for fruit in range(len(s)):
        if s[fruit]=='A':
            Apples+=1
            if Apples >= 1:               
                Oranges-=1
        else:
            Oranges+=1
            if Oranges >= 1:
                Apples-=1
    if Apples>Oranges: 
        ans = 'Apple'
    else:
        ans = 'Orange'
    
    return ans
s1 = 'AO'
s2 = 'AOO'
s3 = 'AAAOOOAAOOO'
s4 = 'AOOAAAO'
s5 = 'OAAO'
print(predict_winner(s1))
print(predict_winner(s2))
print(predict_winner(s3))
print(predict_winner(s4))
print(predict_winner(s5))
```
